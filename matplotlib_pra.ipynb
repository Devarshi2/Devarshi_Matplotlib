```python
x1 = np.arange(0,5)
y1 = np.arange(50,55)
plt.figure(figsize=(15,10))


plt.subplot(2,2,1)
plt.plot(x1,y1,"r^")
plt.xlabel("start rating")
plt.ylabel("Duration")
plt.title("start rating VS Duration")



x2 = x1*x1
plt.subplot(2,2,2)
plt.plot(x1,y2)
plt.xlabel("start rating")
plt.ylabel("Duration")
plt.title("start rating VS Duration")


x3 = np.arange(100,200,2)
y3 = np.arange(500,550)
plt.xlabel("start rating")
plt.ylabel("Duration")
plt.title("start rating VS Duration")

plt.subplot(2,2,3)
plt.plot(x3,y3,'m*')
plt.xlabel("start rating")
plt.ylabel("Duration")
plt.title("start rating VS Duration")


plt.subplot(2,2,4)
plt.plot(x,y,'g--')
plt.xlabel("start rating")
plt.ylabel("Duration")
plt.title("start rating VS Duration")


```
